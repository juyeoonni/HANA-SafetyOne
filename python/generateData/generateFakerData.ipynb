```python
def select_trans_weight_1(age, gender):
    if 20 <= age < 30:
        if gender == 'M':
            return transWeight_20_M_1
        elif gender == 'F':
            return transWeight_20_F_1
    elif 30 <= age < 40:
        if gender == 'M':
            return transWeight_30_M_1
        elif gender == 'F':
            return transWeight_30_F_1
    elif 40 <= age < 50:
        if gender == 'M':
            return transWeight_40_M_1
        elif gender == 'F':
            return transWeight_40_F_1
    elif 50 <= age < 60:
        if gender == 'M':
            return transWeight_50_M_1
        elif gender == 'F':
            return transWeight_50_F_1
    elif age >= 60:
        if gender == 'M':
            return transWeight_60_M_1
        elif gender == 'F':
            return transWeight_60_F_1
    
    return transWeight_20_M_1
```

```python
def gen_cost(costList,weightsList):
    cost=random.choices(costList, weights=weightsList,k=1)[0]
    return cost

def gen_date(start_date,end_date):
    
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y/%m/%d')


def gen_time(cust_start_hour, cust_start_minute, cust_end_hour, cust_end_minute):
    
    start_time_minutes = cust_start_hour * 60 + cust_start_minute
    end_time_minutes = cust_end_hour * 60 + cust_end_minute
    

    random_time_minutes = random.randint(start_time_minutes, end_time_minutes)

    random_hour = random_time_minutes // 60
    random_minute = random_time_minutes % 60

    return f'{random_hour:02d}:{random_minute:02d}:00'

```
